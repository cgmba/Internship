2) In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a data frame and csv. In case if any product has less than 3 pages in search results then scrape all the products available under that product name. Details to be scraped are: "Brand
Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and
“Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.

```python
# Initialize empty lists to store scraped data
brand = []
product = []
price = []
return_exchange = []
delivery = []
availability = []
product_url = []

# Input from the user
search_query = input("Enter the product to search for on Amazon.in: ")
num_pages = 3

driver = webdriver.Chrome()

# Loop through search result pages
for page_num in range(1, num_pages + 1):
    base_url = f"https://www.amazon.in/s?k={search_query}&page={page_num}"
    driver.get(base_url)
    time.sleep(3)  # Wait for the page to load


brand = []

for i in product_url:
    driver.get(i)
    time.sleep(5)
    try:
        brand_tag = driver.find_element(By.XPATH, '/html/body/div[2]/div/div[5]/div[3]/div[4]/div[47]/div/table/tbody/tr[1]/td[2]/span')
        brand.append(brand_tag.text)
    except NoSuchElementException:
        brand.append('Not present')

    try:
        price_tags = driver.find_elements(By.XPATH, '//span[@class="a-price-whole"]')
        for tag in price_tags:
            price_amount = tag.text.strip()
            price.append(price_amount)
    except NoSuchElementException:
        price.append('-')

    try:
        product_tags = driver.find_elements(By.XPATH, '//span[@class="a-size-large product-title-word-break"]')
        for tag in product_tags:
            product_name = tag.text.strip()
            product.append(product_name)
    except NoSuchElementException:
        product.append('-')

    try:
        exchange_tags = driver.find_element(By.XPATH, '/html/body/div[2]/div/div[5]/div[3]/div[4]/div[24]/div[2]/div/div/div/div[2]/div/ol/li[3]/div/span')
        for tag in exchange_tags:
            exchange = tag.text.strip()
            return_exchange.append(exchange)
    except NoSuchElementException:
        return_exchange.append('-')

    try:
        delivery_tags = driver.find_elements(By.XPATH, '/html/body/div[2]/div/div[5]/div[3]/div[4]/div[24]/div[2]/div/div/div/div[2]/div/ol/li[3]/div/span/div[1]')
        for tag in delivery_tags:
            delivery_tag = tag.text.strip()
            delivery.append(delivery_tag)
    except NoSuchElementException:
        delivery.append('-')

    try:
        availability_tags = driver.find_element(By.XPATH, '/html/body/div[2]/div/div[5]/div[3]/div[1]/div[3]/div/div[1]/div/div/div/form/div/div/div/div/div[3]/div/div[5]/div/div[1]/span')
        for tag in availability_tags:
            available = tag.text.strip()
            availability.append(available)
    except NoSuchElementException:
        availability.append('-')

    try:
        product_url_tags = driver.find_elements(By.XPATH, '//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
        for tag in product_url_tags:
            product_link = tag.get_attribute("href")
            product_url.append(product_link)
    except NoSuchElementException:
        product_url.append('-')

    # Sleep for a while before moving to the next page
    time.sleep(2)

# Create a DataFrame
data = {
    "Brand Name": brand,
    "Name of the Product": product,
    "Price": price,
    "Return/Exchange": return_exchange,
    "Expected Delivery": delivery,
    "Availability": availability,
    "Product URL": product_url
}

df = pd.DataFrame(data)

# Save data to a CSV file
df.to_csv(f'{search_query}_search_results.csv', index=False)

# Close the browser
driver.quit()

print("Scraping and saving complete.")
```
