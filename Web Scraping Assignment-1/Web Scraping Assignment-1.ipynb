	3) Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape and make data frame-
	a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.
	b) Top 10 ODI Batsmen along with the records of their team andrating.
	c) Top 10 ODI bowlers along with the records of their team andrating.


```python
# a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.


response = requests.get("https://www.icc-cricket.com/rankings/mens/team-rankings/odi")

	# Parse the HTML content of the page using BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')

	# Find the table containing the team rankings data
table = soup.find('table', class_='table')

	# Initialize empty lists to store data
teams = []
matches = []
points = []
ratings = []


# Iterate through the rows of the table
for row in table.find_all('tr')[1:11]:  # Skip the header row and get the top 10 teams
    columns = row.find_all('td')
    if len(columns) >= 4:
        team_with_country = columns[1].text.strip()  # Get the team name with the country code

        # Use regular expressions to remove the country code and newline characters
        team_name = re.sub(r'\n.*', '', team_with_country)


        match = columns[2].text.strip()
        point = columns[3].text.strip()
        rating = columns[4].text.strip()

        teams.append(team_name)
        matches.append(match)
        points.append(point)
        ratings.append(rating)

# Create a DataFrame
df = pd.DataFrame({'Team': teams,'Matches': matches,'Points': points,'Rating': ratings})

# Display the DataFrame
df
```

5) Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world and make data frame-


i) Headline
ii) Time
iii) News Link


```python
page = requests.get('https://www.cnbc.com/world/?region=world')

soup = BeautifulSoup(page.content)

# Articles in the website
articles = soup.find_all('div', class_="LatestNews-headlineWrapper")

# Extracting Headlines from the article

Headlines = []

# Iterate through the list of articles
for article in articles:
    # Find the <a> element with the class "LatestNews-headline" within the current article
    headline_element = article.find('a', class_='LatestNews-headline')

    # Extract the text from the found <a> element
    headline_text = headline_element.text

    # Add the extracted headline to the 'Headlines' list
    Headlines.append(headline_text)

# Extracting links of the headlines
links = []

# Iterate through the list of articles
for article in articles:
    # Find the first <a> element within the current article
    anchor_element = article.find('a')

    # Extract the value of the 'href' attribute from the found <a> element
    href_attribute = anchor_element['href']

    # Add the extracted link to the 'links' list
    links.append(href_attribute)

#extracting time

time = []

# Iterate through the list of articles
for article in articles:
    # Find the <time> element with the class "LatestNews-timestamp" within the current article
    time_element = article.find('time', class_='LatestNews-timestamp')

    # Extract the text from the found <time> element
    timestamp = time_element.text

    # Add the extracted timestamp to the 'time' list
    time.append(timestamp)

#Create the DataFrame
News = pd.DataFrame({"Headline": Headlines, "Time":time, "links":links})

# Display the DataFrame
News
```

6) Write a python program to scrape the details of most downloaded articles from AI in last 90 days.https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles Scrape below mentioned details and make data frame-
i) Paper Title
ii) Authors
iii) Published Date
iv) Paper URL


```python
page = requests.get('https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles')

# Page Content
soup = BeautifulSoup(page.content)

#Extract Titles
paper_titles = []

# Find all <a> elements with the class "sc-5smygv-0 fIXTHm" in the BeautifulSoup object (soup)
paper_links = soup.find_all('a', class_='sc-5smygv-0 fIXTHm')

# Iterate through the list of found <a> elements
for link in paper_links:
    # Extract the text content from the current <a> element
    link_text = link.text

    # Add the extracted title to the 'paper_titles' list
    paper_titles.append(link_text)

#Extract Author Name
author = []

# Find all <span> elements with the class "sc-1w3fpd7-0 dnCnAO" in the BeautifulSoup object (soup)
author_spans = soup.find_all('span', class_='sc-1w3fpd7-0 dnCnAO')

# Iterate through the list of found <span> elements
for span in author_spans:
    # Extract the text content from the current <span> element
    author_name = span.text

    # Add the extracted author name to the 'author' list
    author.append(author_name)


#Extracting Published date
date = []

# Find all <span> elements with the class "sc-1thf9ly-2 dvggWt" in the BeautifulSoup object (soup)
date_spans = soup.find_all('span', class_='sc-1thf9ly-2 dvggWt')

# Iterate through the list of found <span> elements
for span in date_spans:
    # Extract the text content from the current <span> element
    date_text = span.text

    # Add the extracted date to the 'date' list
    date.append(date_text)

# Extracting All Paper links

paper = []

# Find all <a> elements with the class "sc-5smygv-0 fIXTHm" in the BeautifulSoup object (soup)
paper_links = soup.find_all('a', class_='sc-5smygv-0 fIXTHm')

# Iterate through the list of found <a> elements
for link in paper_links:
    # Get the value of the 'href' attribute from the current <a> element
    href_attribute = link.get('href')

    # Add the extracted URL to the 'paper' list
    paper.append(href_attribute)

#Creating DataFrame
df = pd.DataFrame({'Paper Title':paper_titles, 'Author':author, 'Published Date':date, 'Paper URL':paper})
df
```


7) Write a python program to scrape mentioned details from dineout.co.in and make data frame-
i) Restaurant name
ii) Cuisine
iii) Location
iv) Ratings
v) Image URL


```python
page = requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')

# Page Content
soup = BeautifulSoup(page.content)

# Extracting restaurant name
names = []

# Find all <a> elements with the class "restnt-name ellipsis" in the BeautifulSoup object (soup)
name_links = soup.find_all('a', class_='restnt-name ellipsis')

# Iterate through the list of found <a> elements
for link in name_links:
    # Extract the text content from the current <a> element
    name_text = link.text

    # Add the extracted name to the 'names' list
    names.append(name_text)

#Extracting cuisine
cuisine = []

# Find all <span> elements with the class "double-line-ellipsis" in the BeautifulSoup object (soup)
cuisine_spans = soup.find_all('span', class_='double-line-ellipsis')

# Iterate through the list of found <span> elements
for span in cuisine_spans:
    # Extract the text content from the current <span> element
    span_text = span.text

    # Split the text using the '|' character and get the second part (index 1) which represents cuisine
    cuisine_part = span_text.split('|')[1].strip()

    # Add the extracted cuisine to the 'cuisine' list
    cuisine.append(cuisine_part)

#Extracting Location
location = []

# Find all <div> elements with the class "restnt-loc ellipsis" in the BeautifulSoup object (soup)
location_divs = soup.find_all('div', class_='restnt-loc ellipsis')

# Iterate through the list of found <div> elements
for div in location_divs:
    # Extract the text content from the current <div> element
    div_text = div.text

    # Add the extracted location information to the 'location' list
    location.append(div_text)


#Extracting Ratings

ratings = []

# Find all <div> elements with the class "restnt-rating rating-4" in the BeautifulSoup object (soup)
rating_divs = soup.find_all('div', class_='restnt-rating rating-4')

# Iterate through the list of found <div> elements
for div in rating_divs:
    # Extract the text content from the current <div> element
    div_text = div.text

    # Add the extracted rating to the 'ratings' list
    ratings.append(div_text)


#Extracting Image Urls

images = []

# Find all <img> elements with the class "no-img" in the BeautifulSoup object (soup)
image_elements = soup.find_all('img', class_='no-img')

# Iterate through the list of found <img> elements
for img in image_elements:
    # Get the value of the 'data-src' attribute from the current <img> element
    data_src_attribute = img.get('data-src')

    # Add the extracted image URL to the 'images' list
    images.append(data_src_attribute)

#Creating DataFrame
df = pd.DataFrame({'Restaurant Names':names, 'Cuisine':cuisine, 'Location':location, 'Rating':ratings,'Images':images })
df
```
