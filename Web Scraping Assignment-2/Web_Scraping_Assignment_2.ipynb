Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10 jobs data.
This task will be done in following steps:
1. First get the webpage https://www.shine.com/
2. Enter “Data Analyst” in “Job title, Skills” field and enter “Bangalore” in “enter the location” field.
3. Then click the search button.
4. Then scrape the data for the first 10 jobs results you get.
5. Finally create a dataframe of the scraped data.



**Note: All of the above steps have to be done in code. No step is to be done manually.**

```python
#importing necessary libraries
#!pip install selenium
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
```

```python
driver = webdriver.Chrome()
#opening the shine page on automated chrome browser
driver.get("https://www.shine.com/")
```

```python

#entering designation and location as required in the question
designation = driver.find_element(By.XPATH, "/html/body/div[1]/div[4]/div/div[2]/div[2]/div[1]/form/div/div[1]/ul/li[1]/div/input")
designation.send_keys('Data Analyst')

```

```python
location = driver.find_element(By.XPATH, '/html/body/div[1]/div[4]/div/div[2]/div[2]/div[1]/form/div/div[1]/ul/li[2]/div/input')
location.send_keys("Bangalore")
```

```python
search = driver.find_element(By.XPATH, "/html/body/div[1]/div[4]/div/div[2]/div[2]/div[1]/form/div/div[2]/div/button")
search.click()
```

```python
job_title= []
job_location = []
company_name = []
experience_required = []

#Scrapping Job title from the given results
title_tags = driver.find_elements(By.XPATH, '//h2[@itemprop="name"]')
for i in title_tags[0:10]:
    title = i.text
    job_title.append(title)

#Scrapping Location tags from the given results
location_tags = driver.find_elements(By.XPATH, '//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_locationIcon__zrWt2"]')

for i in location_tags[0:10]:
    location = i.text
    job_location.append(location)

#Scrapping Company tags from the given results
company_tags = driver.find_elements(By.XPATH,'//div[@class="jobCard_jobCard_cName__mYnow"]')

for i in company_tags[0:10]:
    company = i.text
    company_name.append(company)

#Scrapping Experience tags from the given results
experience_tags = driver.find_elements(By.XPATH, '//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_jobIcon__3FB1t"]')


for i in experience_tags[0:10]:
    experience = i.text
    experience_required.append(experience)
```

```python
#creating the dataframe
df = pd.DataFrame({'Title':job_title, 'Location':job_location, 'Company':company_name, 'Experience':experience_required})
df
```

Q4: Scrape data of first 100 sunglasses listings on flipkart.com. You have to scrape four attributes:

To scrape the data you have to go through following steps:

2. Enter “sunglasses” in the search field where “search for products, brands and more” is written and click the search icon
3. After that you will reach to the page having a lot of sunglasses. From this page you can scrap the required data as usual.


image.png

```python
driver = webdriver.Chrome()
#opening the flipkart page on automated chrome browser
driver.get("https://www.flipkart.com/")
```

```python
#entering the product and clicking search as required in the question
product = driver.find_element(By.CLASS_NAME,'Pke_EE')
product.send_keys('sunglasses')
search = driver.find_element(By.CLASS_NAME,"_2iLD__")
search.click()
```

```python
brand = []
product = []
price = []
offer = []

# Define the start and end page numbers
start = 0
end = 3

#Loop through the Pages
for page in range(start,end):

    ##Scrapping brand names from the given results
    name = driver.find_elements(By.XPATH,'//div[@class="_2WkVRV"]')
    for i in name:
        tags = i.text
        brand.append(tags)

    ##Scrapping product names from the given results
    desc = driver.find_elements(By.XPATH,'//a[@class="IRpwTa"]')
    for i in desc:
        prd = i.text
        product.append(prd)

    ##Scrapping price tags from the given results
    Price = driver.find_elements(By.XPATH,'//div[@class="_30jeq3"]')
    for i in Price:
        pc = i.text
        price.append(pc)

    ##Scrapping Offer from the given results
    Offer = driver.find_elements(By.XPATH,'//div[@class="_3Ay6Sb"]')
    for i in Offer:
        off = i.text
        offer.append(off)

    # Find the next button and click on it
    nextbutton = driver.find_element(By.XPATH,'/html/body/div/div/div[3]/div[1]/div[2]/div[12]/div/div/nav/a[11]')
    nextbutton.click()
    # Wait for 3 seconds for the next page to load
    time.sleep(3)

#Getting First 100 products:
Brand = brand[0:100]
Product =product[0:100]
Price = price[0:100]
Offer = offer[0:100]

df = pd.DataFrame({'Brand':Brand, 'Product_Description':Product,'Price':Price, 'Offer':Offer})
df
```

Q5: Scrape 100 reviews data from flipkart.com for iphone11 phone. You have to go the link:
https://www.flipkart.com/apple-iphone-11-black-64-gb/productreviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&marketplace=FLIPKART

image.png

As shown in the above page you have to scrape the tick marked attributes. These are:

Rating
Review summary
Full review
You have to scrape this data for first 100reviews.


**Note: All the steps required during scraping should be done through code only and not manually.**

```python
driver = webdriver.Chrome()

#opening the flipkart page on automated chrome browser
driver.get("https://www.flipkart.com/apple-iphone-11-black-64-gb/product-reviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZE3ENS&sortOrder=MOST_HELPFUL&certifiedBuyer=false&aid=overall")
```

```python
rating = []
summary = []
full_review =[]


start = 0
end = 10

#Iterating through the review section
for review in range(start,end):

    ##Scrapping rating tags from the given results
    rating_tag = driver.find_elements(By.XPATH,'//div[@class="_3LWZlK _1BLPMq"]')
    for i in rating_tag:
        rating.append(i.text)

    #Extract Summary
    summary_tag = driver.find_elements(By.CLASS_NAME,'_2-N8zT')
    for i in summary_tag:
        summary.append(i.text)

    #Extract Full Review
    review_tag = driver.find_elements(By.XPATH,'//div[@class="t-ZTKy"]')
    for i in review_tag:
        full_review.append(i.text)


df = pd.DataFrame({'Rating':rating, 'Review_Summary':summary, 'Full review':full_review})
df
```

```python
driver = webdriver.Chrome()
#opening the flipkart page on automated chrome browser
driver.get("https://www.flipkart.com/")
```

```python
#entering the product and clicking search as required in the question
product = driver.find_element(By.CLASS_NAME,"Pke_EE")
product.send_keys('sneakers')

search = driver.find_element(By.CLASS_NAME,"_2iLD__")
search.click()
```

```python
brand = []
productDescription =[]
price = []

#Define start and end of the page
start= 0
end = 3

for page in range(start,end):

    ##Scrapping brand name from the given results
    name_tag = driver.find_elements(By.XPATH,'//div[@class="_2WkVRV"]')
    for i in name_tag:
        brand.append(i.text)

    ##Scrapping product from the given results
    product = driver.find_elements(By.XPATH, '//a[@class="IRpwTa"]')
    for i in product:
        productDescription.append(i.text)

    ##Scrapping price tags from the given results
    Price = driver.find_elements(By.XPATH, '//div[@class="_30jeq3"]')
    for i in Price:
        price.append(i.text)

    # Click on the next button to move to the next page
    nexts = driver.find_element(By.CLASS_NAME, '_1LKTO3')
    nexts.click()

    # Use a time.sleep() to wait for the next page to load
    time.sleep(3)

Brand = brand[:100]
Product_description = productDescription[:100]
Price = price[:100]

df = pd.DataFrame({'Brand':Brand, 'Product_description':Product_description, 'Price':Price})
df
```

Q8: Write a python program to scrape data for Top 1000 Quotes of All Time.
The above task will be done in following steps:

First get the webpage https://www.azquotes.com/
Click on TopQuotes
Than scrap a) Quote b) Author c) Type Of Quotes



image.png

```python
driver = webdriver.Chrome()
#opening the azquotes page on automated chrome browser
driver.get("https://www.azquotes.com/")
```

```python
#clicking search as required in the question

click = driver.find_element(By.XPATH,"//a[text()='Top Quotes']")
click.click()
```

```python
Quote =[]
Author = []
Type_Of_Quotes = []

#Define Start and end of the pages
start = 0
end = 10

for page in range(start,end):

    #Scrapping quotes from the given results
    quote = driver.find_elements(By.XPATH, '//a[@class="title"]')
    for i in quote:
        Quote.append(i.text)

    #Scrapping author names from the given results
    name = driver.find_elements(By.XPATH, '//div[@class="author"]')
    for i in name:
        Author.append(i.text)

    #Scrapping Type of quotes from the given results
    tags = driver.find_elements(By.XPATH, '//div[@class="tags"]')
    for i in tags:
        Type_Of_Quotes.append(i.text)



df = pd.DataFrame({'Quote':Quote, 'Author':Author, 'Type_Of_Quotes':Type_Of_Quotes})
df
```
